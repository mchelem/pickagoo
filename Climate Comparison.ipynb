```python
"""
Analysis of a few climate indicators for Brazilian cities

Data from the Instituto Nacional de Metereologia (INMET) - from 1961 to 1990
http://www.inmet.gov.br/portal/index.php?r=clima/normaisClimatologicas
"""

from collections import OrderedDict
import pandas
```

```python
# == Scrubbing (helper functions) ==

def read_inmet_table(filepath, UF=None):
    data = pandas.io.parsers.read_csv(filepath, header=3, index_col=1, encoding="UTF-8")

    # Filter by UF
    if UF:
        data = data[data["UF"] == UF]
        
    # Drop non used columns
    del data["UF"]
    del data["Ano"]
    del data[u"Código"]
    
    return data.T


def plot_comparison(data, cities, title, ylabel):
    data[cities].plot(marker='o')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(loc="upper left", bbox_to_anchor=(1,1))
    
    
def plot_min_max(data, title, ylabel):
    max_data = data.max()
    min_data = data.min()
    plot_comparison(data, [max_data.index[max_data.argmax()], min_data.index[min_data.argmin()]], title, ylabel)
```

```python
# == Temperature Analisys ==

# Data containing temperaturesdata[cities].plot(marker='o')
max_temp = read_inmet_table("data/Temperatura-Maxima_NCB_1961-1990.csv")
min_temp = read_inmet_table("data/Temperatura-Minima_NCB_1961-1990.csv")
```

```python
# Cities to be used in the comparison (should be available on the inmet table - see above)
cities_of_interest = [u"Porto Alegre", u"Florianópolis", u"Curitiba"]
```

```python
plot_comparison(max_temp, cities_of_interest, u"Temperatura Máxima", "Temperatura (C)")
```

```python
plot_comparison(min_temp, cities_of_interest, u"Temperatura Mínima", "Temperatura (C)")
```

```python
# == Rain Analysis ==
rain = read_inmet_table("data/Precipitacao-Acumulada_NCB_1961-1990.csv")
rainy_days = read_inmet_table("data/Numero-Dias-Precipitacao-Maior-ou-Igual-1mm_NCB_1961-1990.csv")
```

```python
plot_comparison(rain, cities_of_interest, u"Precipitação", u"Precipitação (mm)")
```

```python
plot_comparison(rainy_days, cities_of_interest, u"Número de dias com precipitação >= 1 mm", u"Número de dias")
```

```python
# == A few extremes ==
plot_min_max(max_temp, u"Temperatura Máxima", "Temperatura (C)")
```

```python
plot_min_max(rain,  u"Precipitação", u"Precipitação (mm)")
```

```python
plot_min_max(rainy_days, u"Número de dias com precipitação >= 1 mm", u"Número de dias")
```
